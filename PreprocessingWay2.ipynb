```python
cleaned_lot_size_units1=[]
default_lot_size_units=data.lot_size_units.mode()[0]
for m in data.lot_size_units:
    if (str(m) == "nan"):
        cleaned_lot_size_units1.append(str(default_lot_size_units))
    else:
        cleaned_lot_size_units1.append(str(m))
```

```python
cleaned_lot_size1=[]
default_lot_size=data.loc[:,'lot_size'].mean()

for s in data.lot_size:
    if str(s) =='nan':
        cleaned_lot_size1.append(float(default_lot_size))
    else:
        cleaned_lot_size1.append(float(s))
for s in cleaned_lot_size1:
    print(s)
```

```python
i=0
for m in data.lot_size_units:
    if (str(m) == "acre"):
        cleaned_lot_size1[i]=float(cleaned_lot_size1[i])*float(43560)
    i=i+1
for s in cleaned_lot_size1:
    print(s)
```

```python
cleaned_df = pd.DataFrame({'beds':data.beds,'baths':data.baths,'size':data.size,'lot_size':cleaned_lot_size1,'price':data.price})
cleaned_df.to_csv('Cleaned_train.csv')
```
